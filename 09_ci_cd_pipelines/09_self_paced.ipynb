Continuous Integration often automates the build process via a "Build Script."  Take a look at the docs for a few examples of build script tools and compare them to the tools we've studied so far:
	
GNU Make: https://www.gnu.org/software/make/manual/html_node/Simple-Makefile.html
		
Apache Ant: https://ant.apache.org/manual/using.html
		
Python setuptools (usually used to create packages to deploy to PyPI):  https://packaging.python.org/en/latest/tutorials/packaging-projects/#creating-the-package-files
	
	
	
Most open source projects make great use of build scripts and build automation, because it makes it far easier for more contributors to participate on a project. For example, many Linux tools, to get the very latest version, require you to "build from source," but this is easier than you might expect due to the build scripts.  Take a look at the process for building the CPython interpreter from source (you'd just need to clone the repo).  Feel free to try out for yourself if you like:  https://github.com/python/cpython#build-instructions
	
While Python code is not "built" in the traditional sense (since it doesn't compile to binary files), there are often cases where you may want to distribute a Python application without making your users install Python themselves, etc.  Check out these Python projects that can aid with that process, and even create cross-platform builds:
	

		
Py2Exe (for Windows builds): https://www.py2exe.org/
		
PyInstaller (cross platform builds): https://pyinstaller.org/en/stable/
